**Balanceamento de Classes**

```python
# Balanceamento de classes usando SMOTE
from imblearn.over_sampling import SMOTE
from collections import Counter

# Separar features e target
X = df_encoded.drop(columns=["Churn_Yes"])  # Ajuste o nome da coluna do target
y = df_encoded["Churn_Yes"]

print("📊 Distribuição original:", Counter(y))

# Aplicar SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

print("📊 Distribuição após SMOTE:", Counter(y_res))
```

**Normalização ou Padronização**

```python
# Flatten nested JSON columns
nested_cols = ['customer', 'phone', 'internet', 'account']

# Create new columns from nested data
for col in nested_cols:
    # Apply pd.json_normalize to flatten each nested column
    # The record_path is None because we want to flatten the dictionary directly
    # The meta is the original columns we want to keep
    df_nested = pd.json_normalize(df[col], record_path=None, meta=[c for c in df.columns if c not in nested_cols])

    # Merge the flattened data back into the original DataFrame
    # We need to make sure the indices align correctly
    df = df.drop(columns=[col]).join(df_nested, lsuffix='_left', rsuffix='_right')

# Drop the original nested columns again, in case any were not dropped in the loop
df = df.drop(columns=nested_cols, errors='ignore')


print("✅ Nested columns flattened and original columns dropped!")
display(df.head())
```
